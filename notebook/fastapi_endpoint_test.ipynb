```python
import sys
import os
import requests

FASTAPI_URL = "https://late-predictor.onrender.com/"
```

### Example of a Prediction Inferencing

```python
url = f"{FASTAPI_URL}/predict"

payload = {
    "datetime_val": "2026-05-06T15:30:00Z",
    "init_latlon": [1.3, 103.8],
    "dest_latlon": [1.35, 103.9],
    "category_id": "c59c0d90-d940-4f0c-a29a-c48c0bf6d9f9"
}

response = requests.post(url, json=payload)

print(response.status_code)
print(response.json())
```
